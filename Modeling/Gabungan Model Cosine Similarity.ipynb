```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
combined_text = df_places['Place_Name'] +  ' ' + df_places['Category'] + ' ' + df_places['City']
combined_text
```

```python
# Menghitung nilai TF-IDF dari deskripsi tempat wisata yang telah di-preprocess
tfidf_matrix = tfidf.fit_transform(combined_text)

# Konversi hasil menjadi DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
tfidf_df
```

```python
df_places = pd.concat([df_places["Place_Id"], tfidf_df], axis=1)
df_places
```

```python
# Menghitung similarity matrix menggunakan cosine_similarity dari scikit-learn
similarity_matrix = cosine_similarity(tfidf_matrix)

# Mengubah similarity matrix menjadi tensor TensorFlow
similarity_tensor = tf.convert_to_tensor(similarity_matrix)
```

```python
# Input place_name
input_place = "Museum Mpu Tantular"

# Get index of input place
place_index = tourism_places[tourism_places["Place_Name"] == input_place].index[0]

# Calculate top_k places based on similarity using tf.nn.top_k
k = 3
top_k_values, top_k_indices = tf.nn.top_k(similarity_tensor[place_index], k=k+1)

# Get top_k places' information
top_k_places = tourism_places.loc[top_k_indices.numpy()]

# Print the top_k places
print(f"Top {k} Places Similar to '{input_place}':")
for i, place in top_k_places.iterrows():
    print("Place Name:", place["Place_Name"])
    print("Description:", place["Description"])
    print("Category:", place["Category"])
    print("Price", place["Price"])
    print("City:", place["City"])
    print("Rating:", place["Rating"])
    print()
```

```python
from flask import Flask, request
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
class Tourism:
    def __init__(self):
        self.tourism_places = None
        self.similarity_tensor = None
    
    def load_data(self, filename):
        self.tourism_places = pd.read_csv(filename)
        
    def filter_data(self):
        self.tourism_places = self.tourism_places.drop(columns = ['Time_Minutes', "Coordinate", 'Lat', 'Long', 'Unnamed: 11', 'Unnamed: 12'])
        
    def TF_IDF(self):
        tfidf = TfidfVectorizer()
        combined_text = self.tourism_places['Place_Name'] +  ' ' + self.tourism_places['Category'] + ' ' + self.tourism_places['City']
        tfidf_matrix = tfidf.fit_transform(combined_text)
        return tfidf_matrix
    
    def calculate_similarity(self, tfidf_matrix):
        similarity_matrix = cosine_similarity(tfidf_matrix)
        self.similarity_tensor = tf.convert_to_tensor(similarity_matrix)
        
    def recommend(self, input_place, k):
        place_index = self.tourism_places[self.tourism_places["Place_Name"] == input_place].index[0]
        top_k_values, top_k_indices = tf.nn.top_k(self.similarity_tensor[place_index], k=k+1)
        top_k_places = self.tourism_places.loc[top_k_indices.numpy()]
        return top_k_places
```

## Alpha Version

```python
from flask import Flask, request
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
app = Flask(__name__)
```

```python
class Tourism:
    def __init__(self):
        self.tourism_places = None
        self.similarity_tensor = None
        self.tfidf_matrix_tourism = None
        self.load_data("D:/Indonesia Tourism Destination/tourism_with_id.csv")
        self.filter_data()
        self.tfidf_matrix_tourism, self.tfidf = self.TF_IDF()
        self.calculate_similarity(self.tfidf_matrix_tourism)
    
    def load_data(self, filename):
        self.tourism_places = pd.read_csv(filename)
        
    def filter_data(self):
        self.tourism_places = self.tourism_places.drop(columns=['Time_Minutes', "Coordinate", 'Lat', 'Long', 'Unnamed: 11', 'Unnamed: 12'])
        
    def TF_IDF(self):
        tfidf = TfidfVectorizer()
        combined_text = self.tourism_places['Place_Name'] + ' ' + self.tourism_places['Category'] + ' ' + self.tourism_places['City']
        tfidf_matrix = tfidf.fit_transform(combined_text)
        return tfidf_matrix, tfidf
    
    def calculate_similarity(self, tfidf_matrix):
        similarity_matrix = cosine_similarity(tfidf_matrix)
        self.similarity_tensor = tf.convert_to_tensor(similarity_matrix)
        
    def recommend(self, input_place, k):
        place_index = self.tourism_places[self.tourism_places["Place_Name"] == input_place].index
        if len(place_index) > 0:
            place_index = place_index[0]
            top_k_values, top_k_indices = tf.nn.top_k(self.similarity_tensor[place_index], k=k+1)
            top_k_places = self.tourism_places.loc[top_k_indices.numpy()]
            return top_k_places
        else:
            input_vector = self.tfidf.transform([input_place])
            similarity_scores = cosine_similarity(input_vector, self.tfidf_matrix_tourism)[0]
            top_k_indices = tf.nn.top_k(similarity_scores, k=k)
            top_k_places = self.tourism_places.loc[top_k_indices.indices.numpy()]
            return top_k_places
    
class Restaurant:
    def __init__(self):
        self.restaurant_places = None
        self.similarity_tensor = None
        self.tfidf_matrix_restaurant = None
        self.load_data("C:/Users/LENOVO/Downloads/Surakarta_Restaurant.csv")
        self.load_data("C:/Users/LENOVO/Downloads/Jakarta_Restaurant.csv")
        self.filter_data()
        self.tfidf_matrix_restaurant, self.tfidf = self.TF_IDF()
        self.calculate_similarity(self.tfidf_matrix_restaurant)

    def load_data(self, filename):
        data = pd.read_csv(filename)
        if self.restaurant_places is None:
            self.restaurant_places = data
        else:
            self.restaurant_places = pd.concat([self.restaurant_places, data], ignore_index=True)
        
    def extract_city(self, address):
        parts = address.split(",")
        for part in reversed(parts):
            part = part.strip()
            if part.lower().startswith("kota ") or part.lower().endswith(" city"):
                city_name = part.replace("Kota", "").replace("City", "").strip()
                return city_name
        return None
    
    def filter_data(self):
        self.restaurant_places = self.restaurant_places.drop(columns="Unnamed: 0")
        self.restaurant_places["City"] = self.restaurant_places["formatted_address"].apply(lambda x: self.extract_city(x))
        self.restaurant_places = self.restaurant_places[["place_id", "name", "Types", "City", "formatted_address", "user_ratings_total", "rating"]]
        self.restaurant_places.fillna('', inplace=True)
        self.restaurant_places.drop_duplicates(subset='place_id', inplace=True)
        self.restaurant_places.reset_index(drop=True, inplace=True)
        
    def TF_IDF(self):
        tfidf = TfidfVectorizer()
        combined_text = self.restaurant_places['name'] +  ' ' + self.restaurant_places['Types'] + ' ' + self.restaurant_places['City'] + ' ' + self.restaurant_places['formatted_address']
        tfidf_matrix = tfidf.fit_transform(combined_text)
        return tfidf_matrix, tfidf
    
    def calculate_similarity(self, tfidf_matrix):
        similarity_matrix = cosine_similarity(tfidf_matrix)
        self.similarity_tensor = tf.convert_to_tensor(similarity_matrix)
    
    def recommend(self, input_place, k):
        place_index = self.restaurant_places[self.restaurant_places["name"] == input_place].index
        if len(place_index) > 0:
            place_index = place_index[0]
            top_k_values, top_k_indices = tf.nn.top_k(self.similarity_tensor[place_index], k=k+1)
            top_k_places = self.restaurant_places.loc[top_k_indices.numpy()]
            return top_k_places
        else:
            input_vector = self.tfidf.transform([input_place])
            similarity_scores = cosine_similarity(input_vector, self.tfidf_matrix_restaurant)[0]
            top_k_indices = tf.nn.top_k(similarity_scores, k=k)
            top_k_places = self.restaurant_places.loc[top_k_indices.indices.numpy()]
            return top_k_places
    
class Hotel:
    def __init__(self):
        self.hotel_places = None
        self.similarity_tensor = None
        self.tfidf_matrix_hotel = None
        self.load_data("C:/Users/LENOVO/Hotel Indonesia.csv")
        self.filter_data()
        self.tfidf_matrix_hotel, self.tfidf = self.TF_IDF()
        self.calculate_similarity(self.tfidf_matrix_hotel)
    
    def load_data(self, filename):
        self.hotel_places = pd.read_csv(filename)
        
    def filter_data(self):
        self.hotel_places = self.hotel_places.drop(columns = ["Lattitute", "Longitude"])
        
    def TF_IDF(self):
        tfidf = TfidfVectorizer()
        combined_text = self.hotel_places['Hotel_Name'] + ' ' + self.hotel_places['Hotel_Category'] + ' ' + self.hotel_places['City'] + ' ' + self.hotel_places['label']
        tfidf_matrix = tfidf.fit_transform(combined_text)
        return tfidf_matrix, tfidf
    
    def calculate_similarity(self, tfidf_matrix):
        similarity_matrix = cosine_similarity(tfidf_matrix)
        self.similarity_tensor = tf.convert_to_tensor(similarity_matrix)
    
    def recommend(self, input_place, k):
        place_index = self.hotel_places[self.hotel_places["Hotel_Name"] == input_place].index
        if len(place_index) > 0:
            place_index = place_index[0]
            top_k_values, top_k_indices = tf.nn.top_k(self.similarity_tensor[place_index], k=k+1)
            top_k_places = self.hotel_places.loc[top_k_indices.numpy()]
            return top_k_places
        else:
            input_vector = self.tfidf.transform([input_place])
            similarity_scores = cosine_similarity(input_vector, self.tfidf_matrix_hotel)[0]
            top_k_indices = tf.nn.top_k(similarity_scores, k=k)
            top_k_places = self.hotel_places.loc[top_k_indices.indices.numpy()]
            return top_k_places
```

```python
tourism = Tourism()
restaurant = Restaurant()
hotel = Hotel()

@app.route('/recommend-tourism', methods=['GET'])
def get_recommend_tourism():
    input_place = request.args.get('place')
    k = int(request.args.get('k'))
    recommendation = tourism.recommend(input_place, k)
    return recommendation.to_json(orient='records')

@app.route('/recommend-restaurant', methods=['GET'])
def get_recommend_restaurant():
    input_place = request.args.get('place')
    k = int(request.args.get('k'))
    recommendation = restaurant.recommend(input_place, k)
    return recommendation.to_json(orient='records')

@app.route('/recommend-hotel', methods=['GET'])
def get_recommend_hotel():
    input_place = request.args.get('place')
    k = int(request.args.get('k'))
    recommendation = hotel.recommend(input_place, k)
    return recommendation.to_json(orient='records')

if __name__ == '__main__':
    app.run()
```
