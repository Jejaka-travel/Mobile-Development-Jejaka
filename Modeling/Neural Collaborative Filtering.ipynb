## Model

```python
df_places = df_place.copy()
```

```python
df_users = df_user.copy()
```

```python
df_targets = df_target.copy()
```

```python
# Pisahkan fitur dan target dari df_places
X_places = df_places.drop(['User_Id', 'Place_Id'], axis=1)
y = df_targets['User_Rating']
```

```python
y
```

```python
# Normalisasi fitur menggunakan Min-Max Scaler
scaler = MinMaxScaler()
X_places_normalized = scaler.fit_transform(X_places)
```

```python
X_places_normalized
```

```python
# Pisahkan data pengguna dari df_user
X_user = df_users.drop(['User_Id', 'Place_Id'], axis=1)
```

```python
# Normalisasi fitur menggunakan Min-Max Scaler
scaler = MinMaxScaler()
X_users_normalized = scaler.fit_transform(X_user)
```

```python
X_users_normalized.shape
```

```python
# Pisahkan data menjadi data latihan dan data pengujian
X_places_train, X_places_test, y_train, y_test = train_test_split(X_places_normalized, y, test_size=0.2, random_state=123)
X_user_train, X_user_test = train_test_split(X_users_normalized, test_size=0.2, random_state=123)

# Jumlah pengguna, jumlah tempat wisata, dan jumlah fitur pengguna
num_users = df_users['User_Id'].nunique()
num_places = df_places['Place_Id'].nunique()
num_user_features = X_user_train.shape[1]
```

```python
num_user_features
```

```python
from keras.models import Sequential
from keras.layers import Embedding, Dense, Concatenate, Input, Flatten
from keras.models import Model

# Input layer untuk pengguna
user_input = Input(shape=(num_user_features,))
user_dense1 = Embedding(num_users, 32)(user_input)
user_dense2 = Flatten()(user_dense1)
user_dense3 = Dense(16, activation='relu')(user_dense2)

# Input layer untuk tempat wisata
place_input = Input(shape=(X_places_train.shape[1],))
place_dense1 = Embedding(num_users, 32)(place_input)
place_dense2 = Flatten()(place_dense1)
place_dense3 = Dense(16, activation='relu')(place_dense2)

# Gabungkan layer pengguna dan tempat wisata
concat = Concatenate()([user_dense3, place_dense3])

# Dense layer tambahan
dense1 = Dense(8, activation='relu')(concat)

# Dense layer terakhir
output = Dense(1, activation='linear')(dense1)
```

```python
# Buat model
model = Model(inputs=[user_input, place_input], outputs=output)

# Kompilasi model
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['mse'])

# Melatih model
model.fit([X_user_train, X_places_train], y_train, epochs=50, validation_data=([X_user_test, X_places_test], y_test))
```

```python
# Melakukan prediksi pada data pengujian
y_pred = model.predict([X_user_test, X_places_test])
y_pred
```

```python
# Menampilkan beberapa contoh prediksi
for i in range(50):
    print('Urutan Ke-', i+1)
    print('Data Asli:', y_test.iloc[i])
    print('Prediksi:', y_pred[i][0])
    print('---')
```

```python
def is_new_user(user_id):
    # Cek apakah user_id terdapat pada data pengguna
    if user_id in df_user['User_Id'].unique():
        return False
    else:
        return True
```

```python
def recommend_places_combined(user_id, bahari=0, budaya=0, cagar_alam=0, pusat_perbelanjaan=0, taman_hiburan=0, tempat_ibadah=0,
                              bandung=0, jakarta=0, semarang=0, surabaya=0, yogyakarta=0, harga_rendah=0, harga_sedang=0, harga_tinggi=0,
                              ave_rating=4, n=10):
    # Cek apakah user_id terdapat pada data pengguna
    if is_new_user(user_id):
        # Rekomendasi untuk pengguna baru
        return recommend_places_new_user(user_id, bahari, budaya, cagar_alam, pusat_perbelanjaan, taman_hiburan, tempat_ibadah,
                              bandung, jakarta, semarang, surabaya, yogyakarta, harga_rendah, harga_sedang, harga_tinggi,
                              ave_rating, n)
    else:
        # Rekomendasi untuk pengguna lama
        return recommend_places(user_id, n)
```

```python
def recommend_places_new_user(user_id, bahari, budaya, cagar_alam, pusat_perbelanjaan, taman_hiburan, tempat_ibadah,
                              bandung, jakarta, semarang, surabaya, yogyakarta, harga_rendah, harga_sedang, harga_tinggi,
                              ave_rating, n):
    # Buat dataframe pengguna baru
    user_data = pd.DataFrame({
        'User_Id': [user_id],
        'Bahari': [bahari],
        'Budaya': [budaya],
        'Cagar_Alam': [cagar_alam],
        'Pusat_Perbelanjaan': [pusat_perbelanjaan],
        'Taman_Hiburan': [taman_hiburan],
        'Tempat_Ibadah': [tempat_ibadah], 
        'Bandung': [bandung],
        'Jakarta': [jakarta],
        'Semarang': [semarang],
        'Surabaya': [surabaya],
        'Yogyakarta': [yogyakarta], 
        'Harga_Rendah': [harga_rendah],
        'Harga_Sedang': [harga_sedang],
        'Harga_Tinggi': [harga_tinggi],
        'Ave_Rating': [ave_rating]
    })

    # Normalisasi fitur pengguna
    user_features = scaler.transform(user_data.drop('User_Id', axis=1))

    # Mengulang data pengguna sebanyak n kali
    user_features_repeated = np.repeat(user_features, repeats=len(X_places_normalized), axis=0)

    # Mendapatkan prediksi rating untuk tempat wisata
    place_predictions = model.predict([user_features_repeated, X_places_normalized])

    # Menambahkan prediksi rating ke dataframe tempat wisata
    df_recommend['Predicted_Rating'] = place_predictions

    # Mengurutkan tempat wisata berdasarkan prediksi rating
    recommended_places = df_recommend[df_recommend['User_Id'] == user_id].sort_values(by='Predicted_Rating', ascending=False).head(n)

    # Menghapus kolom prediksi rating dari dataframe tempat wisata
    #recommended_places = recommended_places.drop('Predicted_Rating', axis=1)

    return recommended_places
```

```python
def recommend_places(user_id, n):
    # Filter data pengguna berdasarkan user_id
    user_data = df_user[df_user['User_Id'] == user_id]
    user_data = user_data.drop(['User_Id', 'Place_Id'], axis=1).drop_duplicates()
    
    # Normalisasi fitur pengguna
    user_features = scaler.transform(user_data)

    # Mengulang data pengguna sebanyak n kali
    user_features_repeated = np.repeat(user_features, repeats=len(X_places_normalized), axis=0)

    # Mendapatkan prediksi rating untuk tempat wisata
    place_predictions = model.predict([user_features_repeated, X_places_normalized])

    # Menambahkan prediksi rating ke dataframe tempat wisata
    df_recommend['Predicted_Rating'] = place_predictions

    # Mengurutkan tempat wisata berdasarkan prediksi rating
    recommended_places = df_recommend[df_recommend['User_Id'] == user_id].sort_values(by='Predicted_Rating', ascending=False).head(n)

    # Menghapus kolom prediksi rating dari dataframe tempat wisata
    #recommended_places = recommended_places.drop('Predicted_Rating', axis=1)

    return recommended_places
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten

# 1. Persiapan Data
features = df.drop('Place_Ratings', axis=1)
labels = df['Place_Ratings']

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# 2. Normalisasi Data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. Bangun Model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Flatten())
model.add(Dropout(0.2))  # Tambahkan layer Dropout dengan tingkat dropout 20%
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))  # Tambahkan layer Dropout dengan tingkat dropout 20%
model.add(Dense(1, activation='linear'))

# 4. Pelatihan Model
model.compile(loss='mean_squared_error', optimizer='adam')

```

```python
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_data=(X_test_scaled, y_test))

# 5. Evaluasi Model
eval_results = model.evaluate(X_test_scaled, y_test)
print("MSE:", eval_results)
```
